## 1. Simulate Diffraction

```
import urllib.request
from synth_xtal.simulator import simulate_diffraction

# Download a sample structure
urllib.request.urlretrieve("https://files.rcsb.org/download/4LZT.pdb", "4LZT.pdb")

# Simulate to 1.5A
simulate_diffraction("4LZT.pdb", "simulated.mtz", d_min=1.5)
```

## 2. Inspecting Data
Let us use `reciprocalspaceship` to investigate the MTZ dataset.

```
import reciprocalspaceship as rs

ds = rs.read_mtz("simulated.mtz")
print(ds.head())
print(f"Total reflections: {len(ds)}")
```
